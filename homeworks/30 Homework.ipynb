### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(-4,4,100)
lower1_x=np.linspace(-4,1,100)
prob=stats.norm.cdf(1,0,1)
print("小於1的機率:",prob)
all_prob=stats.norm.pdf(x,0,1)
lower1_prob=stats.norm.pdf(lower1_x,0,1)
plt.style.use("seaborn")
plt.plot(x,all_prob)
plt.axvline(x=1,lw=.8,ls="--",alpha=1,color="red")
plt.ylim(0,0.5)
plt.fill_between(lower1_x,lower1_prob,interpolate=True, alpha=0.4, label='lower than 1 probability',color="slategrey")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.text(-1.1,0.1,"小於1的機率",fontsize=13)
plt.tight_layout
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
plt.figure(figsize=(6,6))
plt.plot(x,all_prob)
greater1_x=np.linspace(1,4,100)
greater1_prob=stats.norm.pdf(greater1_x,0,1)
prob=stats.norm.cdf(1,0,1)
print("大於1機率:",1-prob)
plt.axvline(x=1,lw=.8,ls="--",alpha=1,color="red")
plt.fill_between(greater1_x,greater1_prob,interpolate=True, alpha=0.4, label='lower than 1 probability',color="slategrey")
lower_neg1_x=np.linspace(-4,-1,100)
lower_neg1_prob=stats.norm.pdf(lower_neg1_x,0,1)
prob=stats.norm.cdf(-1,0,1)
print("小於 -1 的機率:",prob)
plt.axvline(x=-1,lw=.8,ls="--",alpha=1,color="red")
plt.fill_between(lower_neg1_x,lower_neg1_prob,interpolate=True, alpha=0.4, label='lower than 1 probability',color="slategrey")
plt.ylim(0,0.5)
plt.show()
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python
x=6
##標準化
x_normal=(x-2)/2
prob=stats.norm.cdf(x_normal,0,1)
print("標準化後X小於 6 的機率:",prob)
x=np.linspace(-6,10,100)
all_prob=stats.norm.pdf(x,2,2)
lower6_x=np.linspace(-6,6,100)
lower6_prob=stats.norm.pdf(lower6_x,2,2)
plt.figure(figsize=(6,6))
plt.plot(x,all_prob)
plt.axvline(x=6,lw=.8,ls="--",alpha=1,color="red")
plt.fill_between(lower6_x,lower6_prob,interpolate=True, alpha=0.4, label='lower than 1 probability',color="slategrey")
plt.ylim(0,0.22)
plt.show()
```
