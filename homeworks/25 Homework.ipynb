# 目標：
安裝Basemap並且可以使用它繪製地理資訊。這個教程中，我們將涉及地理繪圖module:Basemap。Basemap是Matplotlib的擴展。

# 重點：
Basemap在地圖集上的擴展應用

# 作業：
1.更改Projection的設定繪製出地球儀圖形(經緯度任選)

"Projection"參數為"ortho"，並以NASA寶石藍輸出圖像

"Projection"參數為"Mill"，所得平展位面

"Projection"參數為"Icc"，並以NASA寶石藍輸出

2.查看resolution / 經緯度座標的繪圖精細度

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,10))
m=Basemap(projection="ortho",lat_0=25,lon_0=120,resolution="c")
m.bluemarble(scale=.5)
m.drawcountries(color="w")
plt.show()
```

```python
m=Basemap(projection="mill",llcrnrlon=-180,llcrnrlat=-90,urcrnrlon=180,urcrnrlat=90,resolution="c")
plt.figure(figsize=(10,10))
m.shadedrelief(scale=0.2)
lats = m.drawparallels(np.linspace(-90, 90, 13),labels=[True,],linewidth=.5,color="w")
lons = m.drawmeridians(np.linspace(-180, 180, 13),labels=[0,0,0,1],linewidth=.5,color="w")
m.drawcountries(color="slategrey")
plt.title("BaseMap",fontsize=20)
plt.show()
```

```python
m=Basemap(projection="lcc",lat_0=24,lat_1=25,lon_0=122,resolution=None,width=12000000,height=9000000)
plt.figure(figsize=(10,10))
m.bluemarble(scale=0.2)
m.drawparallels(np.linspace(-90, 90, 13),labels=[True,],linewidth=.5,color="w")
m.drawmeridians(np.linspace(-180, 180, 13),labels=[0,0,0,1],linewidth=.5,color="w")

plt.show()
```

```python
m=Basemap(projection="lcc",lat_0=24,lat_1=25,lon_0=122,resolution=None,width=12000000,height=9000000)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10,10))
m.shadedrelief(scale=0.2)
m.drawparallels(np.linspace(-90, 90, 13),labels=[True,],linewidth=.5,color="w")
m.drawmeridians(np.linspace(-180, 180, 13),labels=[0,0,0,1],linewidth=.5,color="w")
plt.annotate("臺灣",xy=(6000000,4500000),xytext=(6200000,5000000),fontsize=15,color="red",
             arrowprops=dict(arrowstyle="->",color="k"))
#plt.savefig("Taiwan.png",dpi=300)
plt.show()
```
